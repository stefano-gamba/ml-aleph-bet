```python
import math
```

```python
x = [1.0, 2.0]
w = [0.5, -0.5]
b = 0.1
```

```python
def sigmoid(z):
    return 1 / (1 + math.exp(-z))
```

```python
def dot_product(x, w):
    return sum(x_i * w_i for x_i, w_i in zip(x, w))
```

```python
def forward_pass(x, w, b):
    z = dot_product(x, w) + b
    a = sigmoid(z)
    return a, z
```

```python
forward_pass(x, w, b)
```

```python
def backward_pass(x, y, a):
    dz = a - y
    dw = [dz * x_i for x_i in x]
    db = dz
    return dw, db
```

```python
def train(X, Y, epochs, learning_rate):
    w = [0.5, -0.5]
    b = 0.1
    for epoch in range(epochs):
        for x, y in zip(X, Y):
            a, z = forward_pass(x, w, b)
            dw, db = backward_pass(x, y, a)
            w = [w_i - learning_rate * dw_i for w_i, dw_i in zip(w, dw)]
            b -= learning_rate * db
    return w, b
```
