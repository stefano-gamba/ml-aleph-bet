```python
import math
import random
```

```python
X = [[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0]]
k = 2
```

```python
def euclidean_distance(a, b):
    return math.sqrt(sum((x - y) ** 2 for x, y in zip(a, b)))
```

```python
def update_centroids(clusters, X, k):
    centroids = []
    for i in range(k):
        cluster_points = [X[j] for j in range(len(X)) if clusters[j] == i]
        centroids.append([sum(j) / len(cluster_points) for j in zip(*cluster_points)])
    return centroids
```

```python
def compare_centroids(c1, c2):
  return all([euclidean_distance(c1[i], c2[i]) < 0.1 for i in range(len(c1))])
```

```python
def k_means(X, k, max_iters=100):
    centroids = random.sample(X, k)
    for _ in range(max_iters):
        clusters = []
        for x in X:
            distances = [euclidean_distance(x, c) for c in centroids]
            clusters.append(distances.index(min(distances)))
        new_centroids = update_centroids(clusters, X, k)
        if compare_centroids(centroids, new_centroids):
            break
    centroids = new_centroids
    return centroids, clusters
```

```python
centroids, assignments = k_means(X, k)
```

```python
print("Centroids:", centroids, "Assignments:", assignments)
```
