```python
def dot_product(v1, v2):
    return sum(a * b for a, b in zip(v1, v2))
```

```python
def compute_covariance_matrix(X):
    # Calcolare la media di ogni feature
    mu = [sum(feature) / len(X) for feature in zip(*X)]
    
    # Centrare i dati sottraendo la media
    centered_data = [[x - m for x, m in zip(sample, mu)] for sample in X]

    features_transposed = list(zip(*centered_data))
    
    # Calcolare la matrice di covarianza
    covariance_matrix = [[0.0 for _ in range(len(mu))] for _ in range(len(mu))]
    
    for i in range(len(mu)):
        for j in range(len(mu)):
            covariance_matrix[i][j] = dot_product(features_transposed[j], features_transposed[i]) / (len(X) - 1)
    
    return covariance_matrix
```

```python
cov_matrix = compute_covariance_matrix(X)
print("Matrice di covarianza:")
for row in cov_matrix:    
    print(row)
```

```python
def power_iteration(cov_matrix, num_iters=100):
    # Inizializzare un vettore casuale
    b_k = [1.0 for _ in range(len(cov_matrix))]
    
    for _ in range(num_iters):
        # Calcolare il prodotto matrice-vettore
        b_k1 = [sum(cov_matrix[i][j] * b_k[j] for j in range(len(cov_matrix))) for i in range(len(cov_matrix))]
        
        # Calcolare la norma del nuovo vettore
        norm = math.sqrt(sum(x ** 2 for x in b_k1))
        
        # Normalizzare il nuovo vettore
        b_k = [x / norm for x in b_k1]
    
    return b_k
```
